```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer , WordNetLemmatizer
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer , TfidfVectorizer
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"spamCollection.csv" , sep='\\t', names=['label','message'],encoding='utf-8')
```

```python
df = df.replace({'"': ''}, regex=True)
```

```python
df.head()
```

```python
df.shape
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
df1 = df.copy()  
df1['label'] = df1['label'].map({'ham': 0, 'spam': 1})
plt.figure(figsize=(6,4))
sns.countplot(x=df1['label'], palette='coolwarm')
plt.title("📊 Spam vs. Ham Count")
plt.xticks(ticks=[0,1], labels=["Ham", "Spam"])
plt.show()
```

```python
lamitizer = WordNetLemmatizer()
```

```python
train = []
for i in range(len(df)):
    review = re.sub('^[a-zA-Z]' , ' ' , df['message'][i])
    review = review.lower()
    review = review.split()
    review = [lamitizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    train.append(review)
```

```python
print(train[0])
```

```python
cv = CountVectorizer(max_features=5000)
x = cv.fit_transform(train).toarray()
```

```python
y = pd.get_dummies(df['label'])
y.head()
```

```python
y = y.iloc[:,1].values
print(y)
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 0.20,random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mb = MultinomialNB()
mb.fit(train_x,train_y)
pred = mb.predict(test_x)
print(pred)
cm = confusion_matrix(pred,test_y)
print(cm)
acc = accuracy_score(pred,test_y)
acc
```

```python
pred = mb.predict(test_x)
print(pred)
```

```python
cm = confusion_matrix(pred,test_y)
print(cm)
```

```python
acc = accuracy_score(pred,test_y)
acc
```

```python
pickle.dump(cv,open('transform.pkl','wb'))
```

```python
pickle.dump(mb,open('nlp model.pkl','wb'))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(train_x, train_y)

pred_rf = rf.predict(test_x)
print(pred_rf)

cm_rf = confusion_matrix(pred_rf, test_y)
print(cm_rf)

acc_rf = accuracy_score(pred_rf, test_y)
print("Random Forest Accuracy:", acc_rf)
```

# LSTM MODEL

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder

train_x_dense = train_x.toarray() if hasattr(train_x, 'toarray') else train_x
test_x_dense = test_x.toarray() if hasattr(test_x, 'toarray') else test_x
max_length = 100
train_x_padded = pad_sequences(train_x_dense, maxlen=max_length, dtype="float32")
test_x_padded = pad_sequences(test_x_dense, maxlen=max_length, dtype="float32")
encoder = LabelEncoder()
train_y_encoded = encoder.fit_transform(train_y)
test_y_encoded = encoder.transform(test_y)
model = Sequential([
    Embedding(input_dim=train_x_padded.shape[1], output_dim=128, input_length=max_length),
    LSTM(64, return_sequences=False),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(train_x_padded, train_y_encoded, epochs=5, batch_size=32, validation_data=(test_x_padded, test_y_encoded))
pred_lstm = (model.predict(test_x_padded) > 0.5).astype("int32")
print(pred_lstm.flatten())
cm_lstm = confusion_matrix(test_y_encoded, pred_lstm)
print(cm_lstm)

acc_lstm = accuracy_score(test_y_encoded, pred_lstm)
print("LSTM Accuracy:", acc_lstm)
```

#  SVM  MODEL

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', probability=True, random_state=42)
svm.fit(train_x, train_y)
pred_svm = svm.predict(test_x)
print(pred_svm)
cm_svm = confusion_matrix(pred_svm, test_y)
print(cm_svm)

acc_svm = accuracy_score(pred_svm, test_y)
print("SVM Accuracy:", acc_svm)
```
